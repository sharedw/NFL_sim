```python
import numpy as np
import pandas as pd

class KalmanFilterNFL:
    def __init__(self, initial_estimate=0, initial_error=1, process_variance=2, measurement_variance=1):
        self.estimate = initial_estimate
        self.error = initial_error
        self.process_variance = process_variance
        self.measurement_variance = measurement_variance

    def apply_kalman_filter(self, measurements):
        estimates = []
        for measurement in measurements:
            # Prediction step
            self.error += self.process_variance
            
            # Update step
            kalman_gain = self.error / (self.error + self.measurement_variance)
            self.estimate += kalman_gain * (measurement - self.estimate)
            self.error = (1 - kalman_gain) * self.error
            
            estimates.append(self.estimate)
        return estimates

def vectorized_kalman_filter(df, grouper, col):
    initial_estimate = df[col].median()
    def apply_to_group(group):
        kf = KalmanFilterNFL(initial_estimate=initial_estimate, initial_error=1, process_variance=0.1, measurement_variance=2)
        group['kalman_estimate'] = kf.apply_kalman_filter(group[col].to_numpy())
        return group
    
    return df.groupby(grouper, group_keys=False).apply(apply_to_group)
```

```python
data['receptions'].median()
```

```python
stat_col = 'fantasy_points_ppr'

res =vectorized_kalman_filter(data,['player_display_name','player_id'], stat_col).sort_values(by=['season','week'])
res['kalman_estimate'] = res.groupby(['player_id','player_display_name'])['kalman_estimate'].shift(1)
res['mov'] = res.groupby(['player_id','player_display_name'])[stat_col].transform(lambda x: x.rolling(6,min_periods=1).mean().shift(1))
res['kalman_err'] = res[stat_col] - res['kalman_estimate']
res['mov_err'] = res[stat_col] - res['mov']
print(100 *(res['mov_err']**2).mean(), 100 * (res['kalman_err']**2).mean())
res
```
