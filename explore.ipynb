```python
import pandas as pd
```

```python
wk = pd.read_csv('data/weekly.csv', index_col=0)
wk
```

```python
grp = wk.groupby(['season','week','recent_team'])[['receiving_yards','rushing_yards','receptions','carries']].sum()
grp['ypc'] = grp['rushing_yards'] / grp['carries']
grp['ypcatch'] = grp['receiving_yards'] / grp['receptions']
grp['ypc'].quantile([0.03,0.2,0.4,0.5,0.6,0.8,0.99])
```
