```python
depth = nfl.import_depth_charts(years).drop_duplicates().reset_index(drop=True)
```

```python
team_map = {'OAK':'LV',
            'STL':'LA',
            'SD':'LAC'}
depth['club_code'] = depth['club_code'].apply(lambda x: team_map.get(x,x))
```

```python
depth = depth.sort_values(by=['season','club_code','week','depth_team'])
depth['dense_depth'] = depth.groupby(['club_code','season','week','position', 'game_type','formation']).cumcount()
depth['depth_position'] = depth['depth_position'].replace({'\n    ':None})
depth['depth_position'] = depth['depth_position'].fillna(depth['position'])
depth = depth.rename({'club_code':'team_name'},axis=1)
depth
```

```python
depth.to_csv('data/depth.csv')
```

```python
snaps = nfl.import_snap_counts(range(2012,2025))
team_map = {'OAK':'LV',
            'STL':'LA',
            'SD':'LAC'}
snaps['team'] = snaps['team'].apply(lambda x: team_map.get(x,x))
```

```python
len(snaps)
```

```python
#get receiver ids from pbp with name like 'J.Coker' to fill gaps
receivers = pbp[['posteam','season','receiver_player_id', 'receiver_player_name']].dropna().drop_duplicates().copy().rename(
    {'posteam':'team','defteam':'opponent','receiver_player_id':'player_id', 'receiver_player_name':'abbr_name'},axis=1)
receivers['last_name']= receivers['abbr_name'].apply(lambda x: x.split('.')[-1])
#get rosters to fill most of the gaps
rosters = rosters.rename({'player_name':'player'},axis=1)
rosters_ids = rosters.loc[~rosters.player_id.isna()][['player', 'player_id','team','week','season']]
clean_snaps = snaps.merge(ids[['gsis_id','pfr_id','name']], left_on=['pfr_player_id'], right_on=['pfr_id']).dropna()
missing = snaps.merge(ids[['gsis_id','pfr_id','name']],how='left', left_on=['pfr_player_id'], right_on=['pfr_id'])
missing = missing[missing.isna().any(axis=1)].drop(['gsis_id','pfr_id','name'], axis=1)
with_rosters = missing.merge(rosters_ids, how='left').drop_duplicates()
clean_from_rosters = with_rosters.dropna()
#fill remaining gaps with pbp receiver ids from above
still_missing = with_rosters.loc[with_rosters.player_id.isna()].copy().drop(['player_id'],axis=1)
still_missing['last_name'] = still_missing['player'].apply(lambda x: ' '.join(x.split(' ')[1:]))
from_receivers = still_missing.merge(receivers, on=['team','season', 'last_name'], how='left').drop('abbr_name',axis=1)
clean_from_receivers = from_receivers.dropna()
still_missing = from_receivers.loc[from_receivers.player_id.isna()].copy()
```

```python
final = pd.concat((clean_snaps, clean_from_rosters.rename({'player_id':'gsis_id'},axis=1)))
final = pd.concat((final, clean_from_receivers.rename({'player_id':'gsis_id'},axis=1)))
final = final.drop(['last_name','pfr_id','name'],axis=1)
final
```

```python
final.to_csv('data/snap_counts.csv')
```
